```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings 
warnings.filterwarnings('ignore')
```

```python
rcParams['figure.figsize'] = 20,10
rcParams['lines.linewidth'] = 2.5
rcParams['xtick.labelsize'] = 'x-large'
rcParams['ytick.labelsize'] = 'x-large'
```

## Load Dataset

```python
df =pd.read_csv('Clicked Ads Dataset.csv')
df.head(5)
```

## Exploratory Data

### Descriptive Statistics

```python
rows = df.shape[0]
cols = df.shape[1]

print(f'''this dataset has {rows} rows''')
print(f'''this dataset has {cols} columns''')
```

```python
df.info()
```

```python
df1 = df.copy()
df1.drop(columns = ['Unnamed: 0'], inplace = True)
```

```python
df1.isna().sum()
```

```python
num_dtypes = ['int64', 'float64']
nums_df = df1.select_dtypes(include = num_dtypes)
nums = nums_df.columns

print(nums)
```

```python
cat_dtypes = ['object']
cat_df = df1.select_dtypes(include = cat_dtypes)
cat = cat_df.columns

print(cat)
```

```python
df1[nums].describe()
```

```python
df1[cat].describe()
```

```python
for col in cat :
    print(f'''Value count {col} column:''')
    print(df1[col].value_counts())
    print('----------------------------------------------')
    print()
```

### Univariate Analysis

#### Numerical Columns

```python
for i in range(0, len(nums)):
    plt.subplot(2,2,i+1)
    sns.boxplot(x = df1[nums[i]], orient = 'v')
    plt.tight_layout()
```

```python
for i in range(0, len(nums)):
    plt.subplot(2,2,i+1)
    sns.distplot(df1[nums[i]])
    plt.tight_layout()
```

```python
plt.figure(figsize = [15, 4])
num = ['Daily Time Spent on Site', 'Age', 'Daily Internet Usage']

for i in range(0, len(num)):
    plt.subplot(1,3, i+1)
    sns.kdeplot(x = df1[num[i]], hue = df1['Clicked on Ad'])
    # plt.xlabel(num[i])
    # plt.legend(loc = 'upper right')
    plt.tight_layout()
```

```python
plt.figure(figsize = [6,12])

for i in range(0, len(num)):
    plt.subplot(3, 1, i+1)
    sns.boxplot(x = df1[num[i]], y = df1['Clicked on Ad'])
    plt.tight_layout()
```

#### Categorical Columns

```python
plt.figure(figsize=(10, 15))
cat = ['Male','Clicked on Ad', 'city', 'province', 'category']

for i in range(0, len(cat)):
    plt.subplot(3,2,i+1)
    sns.barplot(x=df1[cat[i]].value_counts().index, y=df1[cat[i]].value_counts())
    plt.xticks(rotation = 90)
    plt.tight_layout()
```
